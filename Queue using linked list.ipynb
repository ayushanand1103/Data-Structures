```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class LinkedList: 
    def __init__(self):
        self.head = None

    def add_end(self,data):
        new_node = Node(data)

        if self.head == None:
            self.head = new_node
            return

        curr = self.head
        while curr.next != None:
            curr = curr.next

        curr.next = new_node

    def del_beg(self):
        item = self.head.data
        self.head = self.head.next
        return item
    
    def display(self):
        curr = self.head
        while curr != None:
            print(curr.data)
            curr = curr.next
        
```

```python
class Queue:
    def __init__(self):
        self.q = LinkedList()

    def enqueue(self,data):
        self.q.add_end(data)

    def dequeue(self):
        return self.q.del_beg()
    
    def veiw(self):
        return self.q.display()
```

```python
q = Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

q.veiw()
print("End")
print(q.dequeue())
print(q.dequeue())
```
