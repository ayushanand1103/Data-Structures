```python
class Queue:

    def __init__(self):
        self.queue = []

    def enqueue(self, data):
        self.queue.append(data)

    def dequeue(self):
        return self.queue.pop(0)


```

```python
q = Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
print(q.queue)

print(q.dequeue())
print(q.dequeue())

print(q.queue)
```
