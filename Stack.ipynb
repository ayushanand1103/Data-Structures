```python
class Stack:
    def __init__(self):
        self.values = []

    def push(self, x):
        self.values.append(x)

    def pop(self):
        return self.values.pop()
    
    def peek(self):
        return self.values[-1]
    
    def isEmpty(self):
        if len(self.values) == 0:
            return "Stack is empty"
    
        else :
            return "Stack has some values"
        
    def len(self):
        return len(self.values)
    
    
```

```python
s = Stack()

s.push(10)
s.push(20)
s.push(30)
print(s.len())
print(s.values)
print(s.pop())  
print(s.pop())  
print(s.values)
s.pop()
print(s.isEmpty())
print(s.len())
```
