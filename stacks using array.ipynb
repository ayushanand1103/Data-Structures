```python
import array as arr
```

```python
class stack :
    def __init__(self):
        self.values = arr.array("i",[])

    def push(self , data):
        try:
            self.values.append(data)

        except:
            print("Enter a valid data type")

    def pop(self):
        return self.values.pop()
    
    def peek(self):
        return self.values[-1]
    
    def isEmtpy(self):
        if len(self.values) == 0 :
            x = "Stack is Empty"
        
        else :
            x = "Stack is not Emtpy"
        
        return x

    def top(self):
        return len(self.values) - 1
```

```python
s = stack()
s.push(10)
s.push(20)
s.push(30)
```

```python
print(s.pop())
```

```python
print(s.top())
```
