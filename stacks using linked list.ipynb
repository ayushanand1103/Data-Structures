```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None


```

```python
class linked_list:
    def __init__(self):
        self.head = None
    
    def traverse_print(self):
        curr = self.head 
        while curr != None:
            print(curr.data)
            curr = curr.next

    
    def traverse(self):
        curr = self.head
        if curr == None:
            return None

        while curr.next != None:
            curr = curr.next
            
        return curr.data

    def add_beg(self,x):
        y = Node(x)
        y.next = self.head
        self.head = y

    def delete_beg(self):
        if self.head == None:
            return None
        
        temp = self.head
        self.head = self.head.next
        return temp.data

    def num_values(self):
        curr = self.head
        x = -1
        while curr != None:
            x = x + 1
            curr = curr.next
        return x
```

```python
class stack:
    def __init__(self):
        self.values = linked_list()

    def push(self,x):
        self.values.add_beg(x)

    def pop(self):
        return self.values.delete_beg()
    
    def print_data(self):
        return self.values.traverse_print()
    
    def peek(self):
        return self.values.head.data
    
    def top(self):
        return self.values.num_values()
```

```python
s = stack()

s.push(10)
s.push(20)
s.push(30)

s.values.traverse_print()

print("popped:", s.pop())
```

```python
print(s.top())
```
