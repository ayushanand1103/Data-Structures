```python
import array as arr

class Queue:

    def __init__(self):
        self.queue = arr.array('i')

    def enqueue(self, data):
        self.queue.append(data)

    def dequeue(self):
        item = self.queue[0]
        self.queue.pop(0)
        return item

```

```python

q = Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
print(q.queue)
print(q.dequeue())
print(q.dequeue())

print(q.queue)
```
