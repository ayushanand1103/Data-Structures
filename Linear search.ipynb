```python
def linear_search(arr, target):
    n = len(arr)

    for i in range(n):
        if arr[i] == target:
            return i   

    return -1   


```

```python
arr = [10, 23, 45, 70, 11, 15]
target = 70

result = linear_search(arr, target)

if result != -1:
    print("Element found at index:", result)
else:
    print("Element not found")
```
